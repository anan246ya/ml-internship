```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
digits=load_digits()
#1
print(digits.data.shape)
print(digits.images[0])
```

```python
#2
plt.imshow(digits.images[0])
plt.axis("off")
plt.title(digits.target[0])
plt.show()
```

```python
plt.figure(figsize=(12,4))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(digits.images[i],cmap="gray")
    plt.axis("off")
    plt.title(digits.target[i])
plt.subplots_adjust(wspace=1, hspace=0)
plt.show()
```

```python
#3
plt.figure(figsize=(12,4))
test=digits.images[4]
#horizontal flip
plt.subplot(2,5,1)
plt.imshow(test[:,::-1],cmap="gray")
plt.axis("off")
plt.title('Hflip')

#vertical flip
plt.subplot(2,5,2)
plt.imshow(test[::-1,:],cmap="gray")
plt.axis("off")
plt.title("Vflip")

#crop
plt.subplot(2,5,3)
plt.imshow(test[2:8,4:8],cmap="gray")
plt.axis("off")
plt.title("Crop")

#crop2
plt.subplot(2,5,4)
plt.imshow(test[2:8:2,4:8],cmap="gray")
plt.axis("off")
plt.title("CropCrop")

#colour
pic=test.copy()
pic[-1:4,0:8]=0 #no index err?
plt.subplot(2,5,5)
plt.imshow(pic,cmap="gray")
plt.axis("off")
plt.title("Colour")

#vertical flip
plt.subplot(2,5,6)
plt.imshow(pic[::-1,:],cmap="gray")
plt.axis("off")
plt.title("Vflip on Colour")
```

```python
#4
p4=digits.images[4].flatten()
plt.hist(p4, bins=2)
plt.title("Pixel Value Distribution of 4")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.figure()

p5=digits.images[5].flatten()
plt.hist(p5,bins=17)
plt.title("Pixel Value Distribution of 5")
plt.xlabel("Pixel Value")
plt.ylabel("Frequemcy")
```

```python
#5
import numpy as np
plt.figure(figsize=(12,4))
for i in range(10):
    plt.subplot(2,5,i+1)
    images=digits.images[digits.target==i]
    avgim = images.mean(axis=0)
    plt.imshow(avgim, cmap="gray")
    plt.title(f"Avg {i}")
    plt.axis("off")
```
